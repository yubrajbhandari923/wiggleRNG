```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# IBM Brisbane

```python
df = pd.read_csv('QRNG_ibm_brisbane.txt', header=None)
df.columns = ['data']
brisbane_data = np.array([[int(j) for j in list(i)] for i in df['data'].values])

count = 0

brisbane_entropy = [0] * 100
brisbane_expected_value = [0] * 100

for j in range(100):
    count = 0
    for i in range(20000):
        if brisbane_data[i,j] == 0:
            count+=1
    prob = count/20000
    brisbane_entropy[j] = -((prob * np.log2(prob)) + ((1-prob) * np.log2(1-prob)))
    brisbane_expected_value[j] = 20000 - count


print("Entropy: ", brisbane_entropy)

print("Expected Values: ", brisbane_expected_value)

y = np.arange(100)

plt.plot(y, brisbane_entropy)
plt.xlabel("Qbits")
plt.ylabel("Entropy")
plt.title("Entropy on Qbits")
plt.show()

plt.plot(y, brisbane_expected_value)
plt.xlabel("Qbits")
plt.ylabel("Expected Values")
plt.title("Expected Values of Qubits")
plt.show()

```

```python
brisbane_high_entopy_qiskit = {}
for i in range(100):
    if brisbane_entropy[i] < 0.99:
        brisbane_high_entopy_qiskit[i] = brisbane_entropy[i]

brisbane_df = pd.DataFrame(list(brisbane_high_entopy_qiskit.items()), columns=['Index', 'Entropy'])

print(brisbane_df)
```
