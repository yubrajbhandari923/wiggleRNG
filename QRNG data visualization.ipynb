# IBM_Brisbane

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_histogram(data_path, name):
    decimal_numbers = []

    with open(data_path, 'r') as file:
        for line in file:
            binary_number = line.strip() 
            if binary_number:
                decimal_number = int(binary_number, 2)
                decimal_numbers.append(float(decimal_number))

    plt.hist(decimal_numbers, bins=20000)
    plt.title(f"Histogram of Random Numbers Generated by IBM {name}")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show()
```

```python
plot_histogram("data/QRNG_ibm_brisbane.txt", "Brisbane")
```

```python
plot_histogram("data/QRNG_ibm_kyiv.txt", "Kyiv")
```

```python
plot_histogram("data/QRNG_ibm_sherbrooke.txt", "Sherbrooke")
```

```python
def convert_and_save(input_file, output_file):
    decimal_numbers = []

    with open(input_file, 'r') as file:
        for line in file:
            binary_number = line.strip() 
            if binary_number:
                decimal_number = int(binary_number, 2)
                decimal_numbers.append(decimal_number)

    with open(output_file, 'w') as file:
        for number in decimal_numbers:
            file.write(f"{number}\n")

input_file = "data/QRNG_ibm_brisbane.txt"
output_file = "data/sorted_decimal_numbers.txt"
convert_and_save(input_file, output_file)
```

```python
import numpy as np
import matplotlib.pyplot as plt

file_path = 'sorted_decimal_numbers.txt'

data = np.loadtxt(file_path)

plt.hist(data, bins=20000)

plt.title('Histogram of Data Values')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
